```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x = np.linspace(0, 2 * math.pi, 100)
```

```python
y = np.sin(x);
plt.plot(x, y)
plt.show()
```

```python
def linear_regression(x: list, y: list):
    x_mean = np.sum(x) / len(x)
    y_mean = np.sum(y) / len(y)
    result_x = x - x_mean
    result_y = y - y_mean

    b1 = (result_x * result_y) / result_x **2
    b0 = y_mean - (b1 * x_mean)
    return b0, b1
```

```python
b0, b1 = linear_regression(x, y)
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(x, y, label='Data points', color='blue')
plt.plot(x, b0 + b1 * x, label='Regression line', color='red')
plt.xlabel('Input (x)')
plt.ylabel('Output (y)')
plt.title('Linear Regression on y = sin(x)')
plt.legend()
plt.show()
```
